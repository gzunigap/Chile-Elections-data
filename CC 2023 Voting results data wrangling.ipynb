
To add the election results to a map, we need to add latitude and longitude coordinates to the votation locations
First we make a list with unique values for votation locations. It needs to include the name of School or building used and the region since schools names tend to repeat across the country

```python
# Combine 'Comuna' and 'Local' columns into a list separated by comma and add it as a new column on data df
ubicaciones = pd.DataFrame(data_sum[['Local', 'Comuna']].apply(lambda x: f"{x['Local']}, {x['Comuna']}", axis=1))

data_sum['nombre_local_comuna']=ubicaciones
data_sum.head(5)

#To perform the search for coordinates, we create a separate list with the unique locations
ubicaciones_unicas=ubicaciones.drop_duplicates()
ubicaciones_unicas.columns=['nombre_local_comuna']
ubicaciones_unicas= ubicaciones_unicas.reset_index(drop=True)
```



Now we complete the coordinates using google maps api

```python
import requests
```

```python
def get_lat_long(address):
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address}&key=YOUR-API-KEY" #REPLACE WITH YOUR API KEY WHEN RUNNING THIS CODE
    response = requests.get(url)
    if response.status_code == 200:
        dataa = response.json()
        if dataa['status'] == 'OK':
            location = dataa['results'][0]['geometry']['location']
            return location['lat'], location['lng']
    return None, None

# Apply the function to each address and create new columns for latitude and longitude
latitudes = []
longitudes = []
for address in ubicaciones_unicas['nombre_local_comuna']:
    lat, lng = get_lat_long(address)
    latitudes.append(lat)
    longitudes.append(lng)

ubicaciones_unicas["Latitude"] = latitudes
ubicaciones_unicas["Longitude"] = longitudes
```

The previuos code took a long time to execute. The following code is faster

```python

def batch_geocode(addresses):
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        "address": "|".join(addresses),
        "key": "YOUR-API-KEY"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['status'] == 'OK':
            results = data['results']
            return [(result['geometry']['location']['lat'], result['geometry']['location']['lng']) for result in results]
    return [(None, None)] * len(addresses)  # Return None values if geocoding fails

def get_lat_long(address):
    global cache
    if address not in cache:
        cache[address] = batch_geocode([address])[0]
    return cache[address]

# Initialize a cache dictionary to store already geocoded addresses
cache = {}

# Apply the function to each unique address and create new columns for latitude and longitude
unique_addresses = ubicaciones_unicas['nombre_local_comuna'].unique()
geocoded_results = [get_lat_long(address) for address in unique_addresses]

# Map the geocoded results to the original dataframe
geocoded_dict = dict(zip(unique_addresses, geocoded_results))
ubicaciones_unicas["Latitude"] = ubicaciones_unicas['nombre_local_comuna'].map(lambda x: geocoded_dict[x][0] if geocoded_dict[x] else None)
ubicaciones_unicas["Longitude"] = ubicaciones_unicas['nombre_local_comuna'].map(lambda x: geocoded_dict[x][1] if geocoded_dict[x] else None)

```

Finally we merge the coordinates df with the data_sum df to obtain the total votes for each candidate at a given location including its latitude and longitude

```python
data_final=pd.merge(left=data_sum,right=ubicaciones_unicas,how='left')

data_final.to_csv('datos_final.csv', index=False)
data_final.to_excel('datos_final.xlsx', index=False)
```
